# Plot histograms

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def plot_rgb_histograms(image_path, bins=256):
    # Load image
    img = cv2.imread(image_path)
    if img is None:
        print("Error: Image not found")
        return
    
    # Calculate histograms
    hist_b = cv2.calcHist([img], [0], None, [bins], [0, 256]).flatten()
    hist_g = cv2.calcHist([img], [1], None, [bins], [0, 256]).flatten()
    hist_r = cv2.calcHist([img], [2], None, [bins], [0, 256]).flatten()
    
    # Create figure with 3 vertical subplots
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))
    
    # Plot each channel
    ax1.bar(range(256), hist_r, color='red', alpha=0.7, label='Red Channel')
    ax1.set_title('Red Channel Histogram')
    ax1.set_xlim([0, 255])
    
    ax2.bar(range(256), hist_g, color='green', alpha=0.7, label='Green Channel')
    ax2.set_title('Green Channel Histogram')
    ax2.set_xlim([0, 255])
    
    ax3.bar(range(256), hist_b, color='blue', alpha=0.7, label='Blue Channel')
    ax3.set_title('Blue Channel Histogram')
    ax3.set_xlim([0, 255])
    
    plt.tight_layout()
    plt.show()

# Example usage
plot_rgb_histograms('in_images/1024x768.png') # test.png
```
